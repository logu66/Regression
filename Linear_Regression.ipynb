# **Slice a Data**

```python
Slice=Bodyfat.drop(['Neck','Chest','Density','Abdomen','Hip','Thigh','Knee','Ankle','Biceps','Forearm','Wrist'],axis=1)

Slice
```

```python
x=Slice.iloc[:,0].values.reshape(-1,1)
y=Slice.iloc[:,-1].values.reshape(-1,1)
```

```python
y
```

```python
x
```

```python
plt.scatter(x,y)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

# **Data Spliting**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# **Applying Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
ln=LinearRegression()

ln
```

```python
ln.fit(x_train,y_train)
```

# **Predict**

```python
y_predict=ln.predict(x_test)

y_predict
```

# **Checking**

```python
check=pd.DataFrame(x_test,columns=['Bodyfat'])

check
```

```python
check['actual Weight']=y_test

check
```

```python
check[' Weight Predicted']=y_predict

check
```

# **Visualize the regression**

```python
plt.scatter(x,y,color='red')
plt.plot(x_test,y_predict,color='blue')
plt.show()
```

# **performing metrix (checking any error found)**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test,y_predict)
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
